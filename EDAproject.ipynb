```python
df['Score'].hist()
plt.title('Score Distribution')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns
sns.boxplot(x=df['Score'])
plt.title('Score Outliers')
plt.show()
```

```python
import seaborn as sns
sns.boxplot(x=df['Score'])
plt.title('Score Outliers')
plt.show()
```

```python
sns.countplot(y='State', data=df)
plt.title('Number of Institutes per State')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[['TLR','RPC','GO','OI','PERCEPTION','Score','Rank']].corr(),
            annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
